```python
# importamos las librerías más comunes.
import pandas as pd # pandas nos será útil para la exploración y el manejo de los datos.
import numpy as np # numpy será útil para el cáclulo numércio.
import matplotlib.pyplot as plt
import seaborn as sns # tanto seaborn como matplotlib serán útiles para graficar.
```

```python
# En primer lugar leemos los datos del csv. 
apple = pd.read_csv("../data/aapl.csv", sep=",", header = 1)
apple.info()# Echamos un ojo a la forma que tiene apple
apple #Echamos un ojo a apple
```

```python
# Vamos a modificar ligeramente el data frame de apple para trabajar cómodamente.
# Eliminaremos la columna trimestre, pues con la fecha y los ingresos nos vale.
# También fijaremos la columna "fecha" como ínidice.
apple["fecha"] = pd.to_datetime(apple["fecha"])
apple = apple[["fecha", "Ingresos"]]
apple = apple.set_index("fecha")
apple.head()
```

```python
# Comprobamos que no hay ningún NA en el data frame.
# Lo hacemos sumando los True (valor=1) y False (valor=0) que nos encontramos en la columna de "Ingresos".
# Dado que la suma es 0, todos los valores son False y por ende no hay NA.
sum(apple.isna()["Ingresos"])
```

```python
# Volvemos a echarle un ojo a la forma que tiene apple con las nuevas modificaciones
apple.info()
```

```python
# convertimos los datos a trimestrales
apple_q = apple.resample("q").last().to_period("Q")
apple_q
```

```python
# Calculamos los ingresos anuales de la compañía Apple (nos servirá para más adelante)
apple_suma_anual = apple.resample("Y").sum().to_period("Y")
apple_suma_anual.head()
```

```python
# Graficamos los ingresos anuales
sns.set(rc = {"figure.figsize":(11,4)}) # Definimos el tamaño de la imagen
ax = apple_suma_anual[2:len(apple_suma_anual)-1].plot(marker = 'o', linestyle = '-', xlim=["1990", "2021"])
ax.set_ylabel("Ventas anuales Apple") # Añadimos título al eje y 
ax.set_xlabel("Fechas anuales"); # Añadimos título al eje x
# Guardamos la imagen para poder incluirla en el informe
plt.savefig("Apple_anual_income_time_series.png", dpi=300, bbox_inches='tight')
```

```python
# Calculamos el incremento anual de las ventas de Apple eliminando los datos 1989 y 2021 ya que no están 
# sus trimestres completos
apple_suma_anual_incr = round(apple_suma_anual["1990":"2020"].pct_change(),3)
apple_suma_anual_incr.head()
```

```python
# Graficamos los ingresos anuales
sns.set(rc = {"figure.figsize":(11,4)}) # Definimos el tamaño de la imagen
ax = apple_suma_anual_incr[1:len(apple_suma_anual)].plot(marker = 'o', linestyle = '-', xlim=["1990", "2021"])
ax.set_ylabel("Ventas anuales Apple") # Añadimos título al eje y 
ax.set_xlabel("Fechas anuales"); # Añadimos título al eje x
# Guardamos la imagen para poder incluirla en el informe
plt.savefig("Apple_anual_income_time_series.png", dpi=300, bbox_inches='tight')
```

```python
# Calculamos también los ingresos de Apple en años fiscales (nos servirá para comparar con Barclays) 
apple_q_fiscal = apple_q.asfreq("Q-OCT")
apple_q_fiscal.tail(7)
```

```python
# Calculamos los ingresos anuales para los años fiscales
apple_q_fiscal_anual = apple_q.asfreq("Q-OCT").resample('A-OCT').sum()
apple_q_fiscal_anual.tail(7)
```

```python
################################################# ETS MODEL #################################################
# Nos olvidamos de momento de los años fiscales. Más tarde haremos uso de ellos. Ahora vamos a comenzar con 
# los modelos ETS y ARIMA
```

```python
# Graficamos la serie temporal de los ingresos de Apple
sns.set(rc = {"figure.figsize":(11,4)}) # Definimos el tamaño de la imagen
ax = apple.plot(marker = 'o', linestyle = '-')
ax.set_ylabel("Ventas Apple") # Añadimos título al eje y 
ax.set_xlabel("Fechas"); # Añadimos título al eje x
# Guardamos la imagen para poder incluirla en el informe
plt.savefig("Apple_time_series.png", dpi=300, bbox_inches='tight')
```

```python
# Cargamos las librerías
from sktime.forecasting.base import ForecastingHorizon
from sktime.utils.plotting import plot_series
from sktime.forecasting.model_selection import temporal_train_test_split
```

```python
# Vamos a separar ahora el training del testing para luego comprobar que nuestro modelo sea bueno.
# Vamos a tomar los últimos 8 trimestres como testing
y_train, y_test = temporal_train_test_split(y = apple_q['Ingresos'], test_size = 8)

# we will try to forecast y_test from y_train
# plotting for illustration

plot_series(y_train, y_test, labels = ["y_train","y_test"])
plt.savefig("Training and testing election.png", dpi=300, bbox_inches="tight")
```

```python
# Visualizamos el tamaño del training y del testing
print(y_train.shape[0], y_test.shape[0])
```

```python
from sktime.forecasting.ets import AutoETS
```

```python
fh = np.arange(1, 9)

apple_auto_model = AutoETS(auto=True, sp=4, n_jobs=-1)
apple_auto_model
```

```python
apple_auto_model.fit(y_train)
```

```python
# Mostramos el resultado del modelo ETS
apple_auto_model.summary()
```

```python
# Calculamos las predicciones para el horizonte de planificación marcado
apple_predict = apple_auto_model.predict(fh)
print(apple_predict)
```

```python
#from matplotlib.pyplot import figure
# Realizamos el gráfico de la predicción y el valor real de las ventas de Apple
plot_series(y_train, y_test, apple_predict, labels=["Apple","Apple real","Apple pred"])
#plt.figure(figsize=(11, 4))
#plt.rcParams["figure.figsize"] = (11,4)
```

```python
# Para poder verlo de manera más clara, realizamos el gráfico anterior, pero de 2014 en adelante
plot_series(y_train["2014":], y_test, apple_predict, labels=["Apple","Apple real","Apple pred"])
plt.savefig("How good is the ETS model 2014-.png", dpi=300, bbox_inches="tight")
```

```python
# Analicemos ahora los residuos para ver cuan buenas son estas predicciones.
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
# Calculamos el MAPE
mean_absolute_percentage_error(y_test, apple_predict)
```

```python
from sktime.performance_metrics.forecasting import MeanSquaredError
# Calculamos el MSE
mse = MeanSquaredError()
mse(y_test, apple_predict)
```

```python
# Calculamos el RMSE
rmse = MeanSquaredError(square_root=True)
rmse(y_test, apple_predict)
```

```python
# Consideramos un nuevo horizonte de planificación con el fin de predecir qué ocurrira en lo que
# resta de 2021 y en 2022
fh_ETS_2022 = np.arange(1, 7)
```

```python
# Vamos a considerar ahora todos los datos como conjunto de training para hallar las predicciones hasta 2022
# step 3: specifying the forecasting algorithm
y = apple["Ingresos"].astype("float64").to_period("Q")
apple_auto_model.fit(y)
print(apple_auto_model.summary())
```

```python
# Aplicamos al modelo el nuevo horizonte fh_ETS_2022
apple_predict_2022 = round(apple_auto_model.predict(fh_ETS_2022),3)
print(apple_predict_2022)
```

```python
# Graficamos la predicción de Apple para lo que resta de 2021 y 2022 a partir de 2014
plot_series(y["2014":], apple_predict_2022, labels=["Apple real","Apple pred"])
plt.savefig("Predictions for 2021 and 2022 with ETS.png", dpi=300, bbox_inches="tight")
```

```python
############################################# ARIMA MODEL ########################################################
```

```python
from sktime.transformations.series.boxcox import LogTransformer
```

```python
transformer = LogTransformer()
apple_log = transformer.fit_transform(apple_q)
apple_log.tail()
```

```python
sns.set(rc = {"figure.figsize":(11,4)})
ax = apple_log.plot(marker='o', linestyle='-')
ax.set_ylabel('Ventas Apple')
ax.set_title('Ventas Apple: Transformación logartítmica')
```

```python
from sktime.utils.plotting import plot_correlations
```

```python
plot_correlations(apple_log)
```

```python
from sktime.transformations.series.difference import Differencer
```

```python
# Autocorrelation Fuction
# d=1; D=0; S=4
# 
transf_diff = Differencer(lags = [1])
transf_diff
plot_correlations(transf_diff.fit_transform(apple_log))
```

```python
transf_diff=Differencer(lags=[4])
plot_correlations(transf_diff.fit_transform(apple_log))
```

```python
from sktime.forecasting.base import ForecastingHorizon
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.forecasting.arima import AutoARIMA
```

```python
y_train,y_test = temporal_train_test_split(y=apple_q, test_size=8)
log_y_train,log_y_test = temporal_train_test_split(y=apple_log, test_size=8)
plot_series(y_train, y_test, labels=["Train","Test"])
plot_series(log_y_train, log_y_test,labels=["Log_train","Log_test"])
```

```python
fh_ARIMA = np.arange(1,9)# forecasting horizon fh
fh_ARIMA
```

```python
forecaster = AutoARIMA(sp=4, suppress_warnings=True)
forecaster.fit(log_y_train)
```

```python
print(forecaster.summary())
```

```python
log_y_pred = forecaster.predict(fh_ARIMA)
log_y_pred
```

```python
round(np.exp(log_y_pred),3)
```

```python
plot_series(y_train, y_test, np.exp(log_y_pred), labels=["Apple","Apple real","Apple pred"])
```

```python
plot_series(y_train["2014":], y_test, np.exp(log_y_pred), labels=["Apple","Apple real","Apple pred"])
plt.savefig("How good is the ARIMA model 2014-.png", dpi=300, bbox_inches="tight")
```

```python
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError
```

```python
# Calculamos MAPE para los datos logarítmicos
mean_absolute_percentage_error(log_y_test, log_y_pred)
```

```python
# Calculamos MAPE para los datos reales
round(mean_absolute_percentage_error(y_test, np.exp(log_y_pred)),3)
```

```python
mse = MeanSquaredError()
round(mse(y_test, np.exp(log_y_pred)),3)
```

```python
# Calculamos RMSE para los datos logarítmicos
rmse = MeanSquaredError(square_root=True)
rmse(log_y_test, log_y_pred)
```

```python
# Calculamos RMSE para los datos reales
rmse = MeanSquaredError(square_root=True)
round(rmse(y_test, np.exp(log_y_pred)),3)
```
