```python
# Librerías y funciones generales
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sktime.utils.plotting import plot_series
```

```python
# Modelo ETS
import statsmodels.api as sm
from sktime.forecasting.base import ForecastingHorizon
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.forecasting.ets import AutoETS
import statsmodels.tsa.seasonal as smts
```

```python
# Modelo ARIMA
from sktime.transformations.series.boxcox import LogTransformer
from sktime.utils.plotting import plot_correlations
from sktime.forecasting.arima import AutoARIMA
```

```python
# Medición de errores
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError
```

```python
directory = 'aapl.csv'
aapl_df = pd.read_csv(directory, sep=';', decimal=',', parse_dates=['fecha'], index_col='fecha')
aapl_df = aapl_df.drop(labels='Trimeste', axis='columns').resample('q').last().dropna()
var_ventas = plt.plot(aapl_df.resample('A').sum().pct_change().mul(100))
aapl_df_annual = pd.Series(aapl_df['Ingresos'])
```

```python
plt.style.use('ggplot')
ax = aapl_df.plot(color='blue', linewidth=1, marker='.', markersize=5)
ax.set_title('Evolución de los ingresos de Apple 1989Q4-2021Q2', fontsize=12)
ax.set_xlabel('Fecha', fontsize=9)
ax.set_ylabel('Ingresos (via ventas)', fontsize=9)
plt.show()
```

```python
components = smts.seasonal_decompose(aapl_df['Ingresos'], model='multiplicative', period=4)
components.plot();
```

```python
ax = plt.gca()
sm.graphics.tsa.quarter_plot(aapl_df['Ingresos'],ax=ax)
ax.set_title('Comportamiento Estacional')
plt.show()
```

```python
# test_m = print(int(len(aapl_df['Ingresos'])*0.2))
y_train, y_test = temporal_train_test_split(y = aapl_df['Ingresos'].astype('float64').to_period('Q'), test_size=8)
plot_series(y_train, y_test, labels=['y_train', 'y_test'])
```

```python
fh = np.arange(1, 9)

aapl_auto_model = AutoETS(auto=True, sp=4, n_jobs=-1)
aapl_auto_model.fit(y_train)

print(aapl_auto_model.summary())
```

```python
pred_values = aapl_auto_model.predict(fh)
print(pred_values)

plot_series(y_train, pred_values, y_test, labels=["AAPL_train", "AAPL_pred", "AAPL_test"])
plot_series(y_train['2017Q1':], pred_values, y_test, labels=["AAPL_train", "AAPL_pred", "AAPL_test"])
```

```python
print('MAPE:', mean_absolute_percentage_error(y_test, pred_values))

mse = MeanSquaredError()
print('MSE:', mse(y_test, pred_values))

rmse = MeanSquaredError(square_root=True)
print('RMSE', rmse(y_test, pred_values))

efecto_covid = pred_values['2020Q1']
print('Efecto COVID:', efecto_covid)
```

```python
fh = np.arange(1,7)
y = aapl_df['Ingresos'].astype('float64').to_period('Q')

aapl_auto_model = AutoETS(auto=True, sp=4, n_jobs=-1)
aapl_auto_model.fit(y)

print(aapl_auto_model.summary())
```

```python
pred_values = aapl_auto_model.predict(fh)
print(pred_values)

plot_series(y, pred_values, labels=['AAPL','AAPL_predict'])
```

```python
ventas_acum_ets_2021 = aapl_df['Ingresos']['2021Q1':].sum() + pred_values[:'2021Q4'].sum()
ventas_acum_ets_2022 = pred_values['2022Q1':'2022Q4'].sum()
print('Ventas acumuladas 2021:',ventas_acum_ets_2021, '\nVentas acumuladas 2022:',ventas_acum_ets_2022)
```

```python
transform = LogTransformer()
log_aapl_df = transform.fit_transform(aapl_df)
ax = log_aapl_df.plot(color='blue', linewidth=1, marker='.', markersize=5)
ax.set_title('Logaritmo de los ingresos de Apple 1989Q4-2021Q2', fontsize=12)
ax.set_xlabel('Fecha', fontsize=9)
ax.set_ylabel('Ingresos (ventas acumuladas)', fontsize=9)
plt.show()
```

```python
plot_correlations(log_aapl_df.to_period('Q'))
plot_correlations(aapl_df.to_period('Q'))
```

```python
y_train, y_test = temporal_train_test_split(y = aapl_df.to_period('Q'), test_size=8)
log_y_train, log_y_test = temporal_train_test_split(y = log_aapl_df.to_period('Q'), test_size=8)

plot_series(y_train, y_test, labels=['y_train','y_test'])
print(y_train.shape[0], y_test.shape[0])
```

```python
fh = np.arange(1, 9)

forecaster = AutoARIMA(sp=4, suppress_warnings=True)
forecaster.fit(log_y_train)
```

```python
print(forecaster.summary())
```

```python
log_y_prediction = forecaster.predict(fh)
print(log_y_prediction)

y_prediction = np.exp(log_y_prediction)
plot_series(y_train['2016Q1':], y_test, y_prediction, labels=['AAPL_train','AAPL_test', 'AAPL_prediction'])
```

```python
print('MAPE:', mean_absolute_percentage_error(y_test, y_prediction))

mse = MeanSquaredError()
print('MSE:', mse(y_test, y_prediction))

rmse = MeanSquaredError(square_root=True)
print('RMSE', rmse(y_test, y_prediction))
```

```python
y = log_aapl_df['Ingresos'].astype('float64').to_period('Q')
fh = np.arange(1,7)

forecaster = AutoARIMA(sp=4, suppress_warnings=True)
forecaster.fit(y)

print(forecaster.summary())
```

```python
y_prediction = np.exp(forecaster.predict(fh))

plot_series(np.exp(y)['2016Q1':], y_prediction, labels=['AAPL','AAPL_prediction'])
```
