## Packages

```python
# Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Esta no hace falta: import yfinance as yf
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.tsa.stattools import grangercausalitytests
from statsmodels.tsa.api import VAR
from statsmodels.tsa.base.datetools import dates_from_str
```

```python
# Importamos la libreria de gráficos Seaborn
import seaborn as sns
sns.set(rc={'figure.figsize':(8, 3)})
```

## Datos
Seleccionamos la base de datos macroeconomicos de Statsmodels

```python

mdata = sm.datasets.macrodata.load_pandas().data
mdata
```

### Los datos son trimestrales. Creamos el índice

```python
# 
dates = mdata[['year', 'quarter']].astype(int).astype(str)
quarterly = dates["year"] + "Q" + dates["quarter"]
quarterly = dates_from_str(quarterly)
quarterly
```

### Seleccionamos los datos
### Elegimos el PIB, el Consumo y la Inversión

```python

mdata = mdata[['realgdp','realcons','realinv']]
mdata.index = pd.DatetimeIndex(quarterly)
returns_df = np.log(mdata).diff().dropna()
returns_df
```

## Test de Causalidad de Granger
Hasta 4 retardos

```python

gc_res = grangercausalitytests(returns_df[['realgdp','realcons']], 4)
```

## Modelo VAR

```python
model = VAR(returns_df)
```

## Estimamos  un modelo VAR(2) sin constante

```python
#
results = model.fit(2,trend='n')
results.summary()
```

## Gráfico de los datos

```python
results.plot();
```

## Correlaciones de los residuos

¿Son ruido blanco?

```python
results.plot_acorr();
```

## Predicción
### Es necesario darle los valores iniciales, para la predicción recursiva. 
### ES el orden del VAR


```python
lag_order = results.k_ar
results.forecast(returns_df.values[-lag_order:], 5)
```

```python
results.plot_forecast(10);
```

## Función de Respuesta al Impulso

```python
irf = results.irf(10)
```

```python
irf.plot(orth=False);
```

## Función de Respuesta al Impulso(Individual)

```python
irf.plot(impulse='realcons');
```

## Función de Respuesta al Impulso ACUMULADA

```python
irf.plot_cum_effects(orth=False);
```

## Descomposiciñon del error de predicción

```python
fevd = results.fevd(5)
```

```python
fevd.summary()
```

```python
results.fevd(20).plot();
```

## Selección Automática del VAR

Input: orden máximo

```python
model.select_order(15)
```

## Estimación del Modelo Seleccionado

```python
results = model.fit(maxlags=15, ic='aic')
results.summary()
```

## Predicción

```python
lag_order = results.k_ar
results.forecast(returns_df.values[-lag_order:], 5)
```

```python
results.plot_forecast(10);
```

### Función de Respuesta al Impulso

```python
irf = results.irf(10)
```

```python
irf.plot(orth=False);
```

### Función de Respuesta al Impulso Individual

```python
irf.plot(impulse='realcons');
```

### Función de Respuesta al Impulso ACUMULADA

```python
irf.plot_cum_effects(orth=False);
```

### Descomposición de la Varianza del Error de Predicción

```python
fevd = results.fevd(5)
```

```python
fevd.summary()
```

```python
results.fevd(20).plot();
```

## Análisis de Causalidad de Granger

```python
results.test_causality('realgdp', ['realinv', 'realcons'], kind='f').summary()
```

### Normalidad

```python
results.test_normality().summary()
```
