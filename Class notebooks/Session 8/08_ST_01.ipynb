 # Calculate Daily Logarithmic Return and Squared
 <div class="hr1"></div>

```python
# Log return
# rentabilidad logarítmica
# siempre que se usen tasas de variación para predecir, se usan tasas de variación logarítmicas
# 
data['log_ret'] =np.log(data.Price).diff()
data.tail()
```

```python
# Squared Log return
# Proxy Volatility

data['Squared'] =data['log_ret']**2
data.tail()
```

 ## Testing mean
 <div class="hr1"></div>

```python
# mean zero test
# voy a ver si la media es 0 con el ttest
from scipy import stats
alpha=0.05
t, p =stats.ttest_1samp(data['log_ret'].dropna(),0.0)
print ('Estadítico t:',t)
print ('p-value:',p)

if  p < alpha:
    print (f'Con {alpha} de significación: Rechazamos la hipotesis de media cero.')
else:
    print (f'Con {alpha} de significación: No podemos rechazar la hipotesis de media cero.')
```

 ## ACF y PACF
 ## Ljung-Box Test
 <div class="hr1"></div>

```python
# Autocorrelation Fuction Original Time Series
from sktime.utils.plotting import plot_correlations
```

```python
# Create proxy Volatility 
ibm_vol=data['Squared'].astype('float64').to_period('D').sort_index().dropna()
ibm_vol.tail()
```

```python
# Plot Autocorrelation
plot_correlations(ibm_vol)
# como tenemos un problema de autocorrelación, tenemos un problema de GARCH
```

```python
# import function test Ljung_Box
import statsmodels.api as sm
test_Ljung_Box=sm.stats.acorr_ljungbox
```

```python
#
# testing Ljung- para confirmar si hay problema GARCH o no
# pvalue < alpha -> No Ruido Blanco
test_Ljung_Box(ibm_vol, lags=[10,20,30], return_df=True)
```


 <div class="hr1"></div>
 ## LM test
 <div class="hr1"></div>

```python
# import function test LM-Arch
import statsmodels.api as sm
test_LM_ARCH=sm.stats.diagnostic.het_arch
```

```python
#
# testing LM-ARCH
# pvalue < alpha -> ARCH
lm, lmpval, f, pval, result_int =test_LM_ARCH(data['log_ret'].dropna(), nlags=20, store=True)
print ('Estadístico LM: ', lm)
print ('p-value: ', lmpval)
print ('Estadístico F: ', f)
print ('p-value: ', pval)
```

```python
# https://github.com/bashtage/arch/
# pip install arch

from arch.univariate import ARX
from arch.univariate import ARCH, GARCH
```

```python
# Modelo para la Media un AR(1)
ar = ARX(100*data['log_ret'].dropna(), lags=[1])
print(ar.fit().summary())
```

```python
# MODELO ERRONEO
# ARCH(1)
ar.volatility = ARCH(p=1)
res = ar.fit(update_freq=0, disp="off")
print(res.summary())
# distribution: normal hace referencia a la distribución de los errores
```

```python
# para que sepa que los datos son diarios hay que decirselo
fig = res.plot(annualize="D")
```

```python
# Análisis Residuos
# voy a calcular la z_{t}
mod1_std_resid=(res.resid / res.conditional_volatility).astype('float64').to_period('D').sort_index().dropna()
```

```python
# Plot Autocorrelation de los residuos
plot_correlations(mod1_std_resid)
# los residuos estandarizados deben ser ruido blanco y no pueden tener problemas de autocorrelación
```

```python
# Plot Autocorrelation
plot_correlations(mod1_std_resid**2)
# la z^2 tiene problemas de heterasticidad???
# tiene comportamiento GARCH.
# para que sea correcto, lo ruidos tienen que ser ruido blanco y su cuadrado no puede ser GARCH. Luego, el modelo
# ARCH no es correct!!
```

```python
#
# testing Ljung-Box
# pvalue < alpha -> No Ruido Blanco
test_Ljung_Box(mod1_std_resid**2, lags=[10,20,30], return_df=True)
```

```python
# qqplot
sm.qqplot(mod1_std_resid,line='45')
plt.show()
# me muestra la representación de la distribución en función de la normal
# esto es, la distribución de los errores no es normal
```

 ## GARCH(1,1)
 <div class="hr1"></div>

```python
# estimamos un GARCH(1,1)
ar = ARX(100*data['log_ret'].dropna(), lags=[1])
print(ar.fit().summary())
```

```python
# MODELO 
# GARCH(1,1)
ar.volatility = GARCH(p=1,q=1)
res = ar.fit(update_freq=0, disp="off")
print(res.summary())
```

```python
#
fig = res.plot(annualize="D")
```

```python
# Análisis Residuos
mod1_std_resid=(res.resid / res.conditional_volatility).astype('float64').to_period('D').sort_index().dropna()
```

```python
# Plot Autocorrelation
plot_correlations(mod1_std_resid)
```

```python
# Plot Autocorrelation
plot_correlations(mod1_std_resid**2)
# veamos si la z^2 es ruido blanco. Ahora vemos que no es problema GARCH y, la z cumple lo que queríamos
```

```python
#
# testing Ljung-Box
# pvalue < alpha -> No Ruido Blanco
test_Ljung_Box(mod1_std_resid**2, lags=[10,20,30], return_df=True)
```

```python
# qqplot
sm.qqplot(mod1_std_resid,line='45')
plt.show()
# el qqplot sigue sin ser normal.
```

```python
# el modelo es correcto, pero no es correcta la distribución. Cambiamos de distribución
```

```python
#
from arch.univariate import StudentsT
```

```python
#
ar.distribution = StudentsT()
res = ar.fit(update_freq=0, disp="off")
print(res.summary())
```

```python
# Análisis Residuos
mod1_std_resid=(res.resid / res.conditional_volatility).astype('float64').to_period('D').sort_index().dropna()
```

```python
# Plot Autocorrelation
plot_correlations(mod1_std_resid)
```

```python
# Plot Autocorrelation
plot_correlations(mod1_std_resid**2)
```

```python
#
# testing Ljung-Box
# pvalue < alpha -> No Ruido Blanco
test_Ljung_Box(mod1_std_resid**2, lags=[10,20,30], return_df=True)
```

```python
# qqplot normal
sm.qqplot(mod1_std_resid,line='45')
plt.show()
```

```python
# qqplot t-student
sm.qqplot(mod1_std_resid,dist=stats.t,fit=True,line='45')
plt.show()
# casi casi lo hemos conseguido.
```

```python
# Predicción 20 días
forecasts = res.forecast(horizon=20, reindex=False)
```

```python
# Mean
forecasts.mean
# se hace constante porque revierte a la media
```

```python
# Varianza
# varianza^(1/2). Para obtener la diaría, necesito multiplicar por la raiz e 252.
(forecasts.variance**0.5)*np.sqrt(252)
# cuanto mas dias pasen, más riesgo, luego más volatilidad.
```
